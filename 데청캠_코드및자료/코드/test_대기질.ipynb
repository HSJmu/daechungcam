## 분석하기 - 서울시 대기질 데이터

```python
data_raw_avg = data_raw.loc[data_raw['Gu']=="평균"]
```

```python
data_raw_avg
```

```python
data_raw_avg= data_raw_avg.sort_values('DateTime')
data_raw_avg
```

```python
# index 다시 지정 0부터
data_raw_avg=data_raw_avg.reset_index(drop=True)
```

### 모든 날짜에 대한 일평균 PM10,PM2.5를 계산하세요.

```python
data_raw_avg['yyyymmdd'].unique()
```

```python
data_raw_avg_group=data_raw_avg.groupby(['yyyymmdd']).mean()
```

```python
data_raw_avg_group.head()
```

#### 미세먼지 농도별 예보등급 구분, pm10 151이상 : 매우나쁨, 81~150 : 나쁨, 
####                                31~80 : 보통, 0~30 : 좋음


```python
condition=[]

for i in data_raw_avg_group['PM10']:
    if i >= 151 :
        condition.append("VB")
    elif i>=81 :
        condition.append("B")
    elif i>=31 :
        condition.append("N")
    else:
        condition.append("G")
```

```python
data_raw_avg_group['condition']=condition
```

```python
data_raw_avg_group
```

```python
data_raw_avg_group['yyyymm'] = data_raw_avg_group['datetime'].dt.strftime("%Y%m")
data_raw_avg_group['day'] = data_raw_avg_group['datetime'].dt.strftime("%d")
```
